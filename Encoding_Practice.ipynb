```python
import pandas as pd
df = pd.DataFrame({'Color':['Yellow','Green','Red','Green'],
             'Level':['Medium','Easy','Difficult','Easy'],
             'Target':[0,1,1,0]})
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Color_Encoded_0'] = le.fit_transform(df.Color)
```

```python
df
```

```python
df['Level_Encoded_0'] = pd.factorize(df['Level'])[0]
df
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder()
df1 = ohe.fit_transform(df['Color'])
df1
```

```python
ohe = OneHotEncoder()
df1 = pd.DataFrame(ohe.fit_transform(df['Color'].values.reshape(-1,1)).toarray(), columns = ['Color_Green','Color_Red','Color_Yellow'])
df1
```

```python
pd.concat([df, df1], axis=1)
```

```python
pd.get_dummies(df, prefix='Color', columns=['Color'])
```

```python
pd.get_dummies(df, columns=['Color'])
```

```python
Level_Dict = {'Easy' : 1, 'Medium' : 2, 'Difficult' : 3}
df['Lever_Ord'] = df.Level.map(Level_Dict)
df
```

```python
df.groupby(['Level']).mean()
```

```python
mean_encode = df.groupby(['Level'])['Target'].mean()
mean_encode
```

```python
df['Level_Mean_Encode'] = df['Level'].map(mean_encode)
df
```
